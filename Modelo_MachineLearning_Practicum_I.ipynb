```python
# Crear un diccionario para mapear nombres de jugadores a IDs
ids_de_jugadores = {}

# Iterar a través de la lista de jugadores y obtener los IDs del diccionario
for jugador in jugadores:
    if jugador in dicc:
        ids_de_jugadores[jugador] = dicc[jugador]
    else:
        ids_de_jugadores[jugador] = "NA"  # Puedes manejar el caso de jugadores no encontrados

ids_de_jugadores
# Imprimir los IDs de los jugadores
#for jugador, id_jugador in ids_de_jugadores.items():
#    print(f"{jugador}: {id_jugador}")
```

```python
def crear_df_por_id(jugador_id, df, ids_de_jugadores):
    # Obtener el nombre del jugador a partir del diccionario de IDs
    jugador = None
    for nombre, id in ids_de_jugadores.items():
        if id == jugador_id:
            jugador = nombre
            break
    
    if jugador is None:
        raise ValueError(f"No se encontró un jugador con el ID {jugador_id}.")
    
    # Filtrar las filas donde el jugador aparece como ganador o perdedor
    mask = (df['winner_id'] == jugador_id) | (df['loser_id'] == jugador_id)
    resultados = df[mask].copy()  # Usar .copy() para evitar una vista del DataFrame original
    
    # Crear la columna 'y' basada en si el jugador es el ganador utilizando .loc
    resultados['y'] = 0  # Inicializar con 0
    resultados.loc[resultados['winner_id'] == jugador_id, 'y'] = 1
    
    # Resetear el índice del DataFrame
    resultados.reset_index(drop=True, inplace=True)
    
    return resultados



crear_df_por_id(134457, df, ids_de_jugadores)


```

```python
resultados_por_jugador = {}  # Un diccionario para almacenar los resultados por jugador
ids_no_guardados = []  # Una lista para rastrear los IDs que no se guardaron

for jugador_id in ids_de_jugadores.values():
    resultados_jugador = crear_df_por_id(jugador_id, df, ids_de_jugadores)
    
    # Verificar si el DataFrame está vacío y si no todos los valores de "y" son ceros
    if not resultados_jugador.empty and (resultados_jugador['y'] != 0).any():
        resultados_por_jugador[jugador_id] = resultados_jugador
    else:
        ids_no_guardados.append(jugador_id)

# resultados_por_jugador ahora contendrá los DataFrames de resultados para jugadores no vacíos
# La lista ids_no_guardados contendrá los IDs que no se guardaron
# resultados_por_jugador ahora contendrá los DataFrames de resultados para jugadores no vacíos

#resultados_por_jugador = {}  # Un diccionario para almacenar los resultados por jugador
#
#for jugador_id in ids_de_jugadores.values():
#    resultados_jugador = crear_df_por_id(jugador_id, df, ids_de_jugadores)
#    resultados_por_jugador[jugador_id] = resultados_jugador
#
# resultados_por_jugador ahora contendrá los DataFrames de resultados para cada jugador
resultados_por_jugador
```
