```python
def crear_df(jugador, df):
    df_w = df['winner_name'] == jugador
    df_l = df['loser_name'] == jugador
    filtro = df_w | df_l
    resultados = df[filtro]
    resultados['y'] = np.where((resultados['winner_name'] == jugador), 1, 0)
    return resultados

crear_df('Carlos Alcaraz', df)
```

```python
# Crear DataFrames para cada jugador y asignarlos a variables
jugador_dataframes = {}
for jugador in jugadores:
    jugador_dataframes[jugador] = crear_df(jugador, df)
    
jugador_dataframes
```

```python
for jugador in jugadores:
    # Obtener el DataFrame correspondiente al jugador actual
    df_jugador = jugador_dataframes[jugador]
    
    # Crear un gráfico de barras para la columna 'y' en el DataFrame del jugador
    plt.figure()  # Crear una nueva figura para cada jugador
    plt.bar(df_jugador['y'].unique(), df_jugador['y'].value_counts())
    plt.title(f'Gráfica de barras para {jugador}')
    plt.xlabel('Valor de y')
    plt.ylabel('Recuento')
    plt.xticks(df_jugador['y'].unique())  # Establecer las etiquetas del eje X a los valores únicos de 'y'
    plt.show() 
```
